# Analisando avaliações do Movie Lens

```python
notas = pd.read_csv("ratings.csv")
notas.head()
```

```python
nota_media_por_filme = notas.groupby("movieId")["rating"].mean()
nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme.values)
ax.set(xlabel = "Nota média", ylabel= "Frequência", 
       title = "Nota média por filme")
```

```python
notas.head()
```

```python
quantidade_votos_por_filme = notas.groupby("movieId").count()
filmes_com_pelo_menos_10_votos = quantidade_votos_por_filme.query("userId >= 10").index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelos_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos]
```

```python
ax = sns.histplot(nota_media_dos_filmes_com_pelos_menos_10_votos.values, kde=True)
ax.set(xlabel = "Nota média", ylabel= "Distribuição", 
       title = "Nota média por filme")
```

```python
ax = sns.boxplot(x = nota_media_dos_filmes_com_pelos_menos_10_votos.values, width=0.2)
ax.set(title = "Distribuição de nota média em filme no Movielens 100k com mais de 10 votos", xlabel = "Nota média")
```

```python
ax = sns.distplot(nota_media_dos_filmes_com_pelos_menos_10_votos, 
                  hist_kws = {'cumulative':True}, 
                  kde_kws = {'cumulative':True})
ax.set(xlabel='Nota média', ylabel='Proporção acumulada de filmes')
ax.set_title('Média de votos em filmes no MovieLens')
```

```python
from statsmodels.stats.weightstats import ztest
ztest(notas1.rating, value = 3.4320503405352594)

# p-value = 8.978190401885648e-18
```

O filme Toy Story tem uma média diferente da média de todos os filmes do dataset

```python
np.random.seed(75241)

temp = notas1['rating'].sample(frac=1)

def calculates_test(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i],value = 3.4320503405352594)

    return (i,media, p)

valores = np.array([calculates_test(i) for i in range(2,len(temp))])

#medias
plt.plot(valores[:,0],valores[:,1])
plt.plot(valores[:,0],valores[:,2])
plt.hlines(y=0.05, xmin = 2, xmax=len(temp), colors="r")
```

```python
zconfint(notas1.rating, notas.rating)
```

O teste mostra a diferença de valores entre a notas1 e notas, tendo como base a notas1, ou seja, a media de notas1 é (0.2799245129192442, 0.5588219849631111) maior que a média de notas
