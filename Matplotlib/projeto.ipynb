### Ano de 2014

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.68,0.56,0.3,0.4])

ax1.plot(df['data'], df['temperatura'], color = "steelblue")
ax1.set_xlim(dt.datetime(2014,1,1), dt.datetime(2015,1,1))
ax1.grid(True)
ax1.set_title("Temperatura em 2014", fontsize = 20, pad= 20)
ax1.legend(['Temperatura'], loc = "upper left", fontsize = 14)
ax1.set_ylabel('Temperatura(K)', fontsize = 16)
ax1.set_xlabel('Data', fontsize = 16)

ax2.plot(df['data'], df['temperatura'], color = "slateblue" , lw = 0.8)
ax2.grid(True)
ax2.set_title("Temperatura no momento", fontsize = 20, pad= 20)
ax2.legend(['Temperatura'], fontsize = 12)
ax2.set_ylabel('Temperatura(K)', fontsize = 10)
ax2.set_xlabel('Data', fontsize = 10)
```

### Mês de Maio

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.58,0.65,0.4,0.3])

ax1.plot(df['data'], df['temperatura'], color = "steelblue")
ax1.set_xlim(dt.datetime(2014,5,1), dt.datetime(2014,6,1))
ax1.set_ylim(270,320)
ax1.grid(True)
ax1.set_title("Temperatura no mês de Maio/2014", fontsize = 20, pad= 20)
ax1.legend(['Temperatura'], loc = "upper left", fontsize = 14)
ax1.set_ylabel('Temperatura(K)', fontsize = 16)
ax1.set_xlabel('Data', fontsize = 16)

azul_esquerda = df['data'] < dt.datetime(2014,5,1)
azul_direita = df['data'] > dt.datetime(2014,6,1)

ax2.plot(df['data'], df['temperatura'], color = "steelblue" , lw = 0.8)
ax2.plot(df[azul_esquerda]['data'], df[azul_esquerda]['temperatura'], color = 'slateblue' )
ax2.plot(df[azul_direita]['data'], df[azul_direita]['temperatura'], color = 'slateblue' )
ax2.set_title("Temperatura em 2014", fontsize = 10)
ax2.set_xlim(dt.datetime(2014,1,1), dt.datetime(2015,1,1))
ax2.grid(True)
ax2.legend(['Temperatura'], fontsize = 12)
ax2.set_ylabel('Temperatura(K)', fontsize = 10)
ax2.set_xlabel('Data', fontsize = 10)
```

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'steelblue')
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

eixo.annotate("Máximo", xy = (x1,y1), fontsize=20, xytext = (x2,y2), arrowprops = dict(facecolor = 'k') )

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 7000]
y2 = min(df['temperatura']) + 5

eixo.annotate("Mínimo", xy = (x1,y1), fontsize=20, xytext = (x2,y2), arrowprops = dict(facecolor = 'k') )

eixo.axhline(max(df['temperatura']), color = "k", linestyle = "--")
eixo.axhline(min(df['temperatura']), color = "k", linestyle = "--")
```

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='black')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
         autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```
