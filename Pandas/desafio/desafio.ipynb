```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace = True)
```

```python
df
```

```python
df.drop(df.query("Nome == 'Alice' | Nome == 'Carlos'").index, axis = 0, inplace = True)
```

```python
df.query("Nome == 'Alice' | Nome == 'Carlos'")
```

```python
df.replace(to_replace = 7, value = 8, inplace = True)
```

```python
df_aproved = df.query("Aprovado == True")
df_aproved
```

```python
df_aproved.to_csv("alunos_aprovados.csv", index=False)
```

# Segundo Desafio

```python
df['Pontos Extras'] = 0.4 * df['Notas']
df.head()
```

```python
df['Notas Finais'] = df['Notas'] + df['Pontos Extras']
df.head()
```

```python
df['Aprovados finais'] = df['Notas Finais'].apply(lambda x: True if x >= 6 else False) 
df.head()
```
