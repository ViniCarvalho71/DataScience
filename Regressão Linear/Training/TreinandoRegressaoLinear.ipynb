```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv("Student_Performance.csv")
```

```python
dados.head(10)
```

```python
dados.columns
```

```python
dados.isnull().value_counts()
```

```python
dados['Extracurricular Activities'] = np.where(dados['Extracurricular Activities'] == 'Yes', 1, 0)
```

```python
dados.describe().round(4)
```

```python
dados.corr().round(4)
```

```python
#Paleta
sns.set_palette("Set3")

#Tema
sns.set_theme(style = "whitegrid")
```

```python
ax = sns.histplot(data = dados['Performance Index'], kde = True)
ax.figure.set_size_inches(12, 8)
ax.set_ylabel("Alunos",fontsize = 12)
ax.set_xlabel("Performance",fontsize = 12)
ax.set_title("Histograma de performance")
```

```python
ax = sns.boxplot(data = dados['Performance Index'], width=0.2)
ax.figure.set_size_inches(12,8)
ax.set_xlabel("Performance",fontsize = 12)
ax.set_title("Boxplot de performance")
```

```python
ax = sns.boxplot(data = dados,y='Performance Index',x='Extracurricular Activities', width = 0.2)
ax.figure.set_size_inches(12,8)
ax.set_xlabel("Atividade",fontsize = 12)
ax.set_ylabel("Performance",fontsize = 12)
ax.set_title("Performance X Atividades extracurriculares")
```

```python
ax = sns.boxplot(data = dados,y='Performance Index',x='Sample Question Papers Practiced', width = 0.5)
ax.figure.set_size_inches(12,8)
ax.set_xlabel("Estudo",fontsize = 12)
ax.set_ylabel("Performance",fontsize = 12)
ax.set_title("Performance X Estudo")
```

```python
ax = sns.boxplot(data = dados,y='Performance Index',x='Sleep Hours', width = 0.5)
ax.figure.set_size_inches(12,8)
ax.set_xlabel("Sono",fontsize = 12)
ax.set_ylabel("Performance",fontsize = 12)
ax.set_title("Performance X Sono")
```

```python
ax = sns.pairplot(data = dados,
                  y_vars = ['Performance Index'],
                  x_vars = ['Previous Scores', 'Extracurricular Activities','Sleep Hours',
                            'Sample Question Papers Practiced', 'Performance Index'], kind = 'reg' )
ax.fig.suptitle('Verificando variáveis explicativas', fontsize=18, y=1.05)
ax.figure.set_size_inches(10,4)
```

```python
ax = sns.jointplot(x="Previous Scores", y="Performance Index", data=dados, kind="reg")
ax.fig.suptitle('Dispersao - Performance X Pontos anteriores', fontsize=18, y=1.05)
ax.set_axis_labels("Pontos anteriores", "Performance", fontsize=14)
ax
```
